# VECTOR, MATRIX,TENSOR

# ARRAY RESHAPING

```python
arr=np.arange(12)
print(arr)
print(arr.shape)

reshaped=arr.reshape((3,4))
print(reshaped)

flattened=reshaped.flatten()
print("\n",flattened)

raveled=reshaped.ravel()
print(raveled)

transpose=reshaped.T
print("\n",transpose)
```
