## SLICING ##

```python
b=[1,2,3,4,5,6,7,8,9,99]
print(b[1:4])
```

```python
print(b) # PRINTING B 
```

```python
print(b[-4:]) # WHEN IN SLICING ENDING RANGE IS NOT GIVEN
```

```python
print(b[-4:-1]) #WHEN IN SLICING BOTH NEGATIVE RANGES ARE GIVEN
```

```python
print(b[:-1]) #WHEN IN SLICING STARTING RANGE IS NOT GIVEN 
```

```python
list = ['p', 'r', 'o', 'g', 'r', 'a', 'm'] #WHEN NO RANGE IS GIVEN 
print(list[:])
```

```python
print(list[:-4])
```

## ADD ELEMENTS ##

```python
fruits=["apple","banana","kiwi"]          #USING APPEND STATEMENTS
fruits.append("orange")
```

```python
print(fruits)
```

```python
fruits=["apple","banana","kiwi"]
fruits.insert(2, 'hamza')                 #USING INSERT STATEMENT at specific index
print(fruits)
```

## CHANGE ELEMENTS ##

```python
fruits[2]="mango"
print(fruits)
```

## remove one element from  a list ##

```python
fruits.remove("kiwi")
print(fruits)                         #REMOVE
```
