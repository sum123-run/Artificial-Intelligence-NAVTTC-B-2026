# TASK 8 #

### FIND AREA OF TRIANGLE WITH 3 SIDES ARE KNOWN ###

```python
len_a = float(input("Enter length of first side: "))
len_b = float(input("Enter length of second side: "))
len_c = float(input("Enter length of third side: "))

# Triangle validation
if (len_a + len_b > len_c) and (len_a + len_c > len_b) and (len_b + len_c > len_a):

    s = (len_a + len_b + len_c) / 2   # semi-perimeter

    area = (s * (s - len_a) * (s - len_b) * (s - len_c)) ** 0.5

    print(f"Area of lengths {len_a}, {len_b}, {len_c} is: {area}")

else:
    print(" These lengths cannot form a triangle")
```

### SOLVE THE QUADRATIC EQUATION ###

```python
import math
a=float(input("enter first number"))
b=float(input("enter your second number"))
c=float(input("enter your third number"))
if (a==0):
    print("it is not quadratic equation because a should not be zero")
else:
    d=(b*b)-4*a*c
    print("Discriminant is :",d)
    if(d>0):
        x1= (-b+math.sqrt(d)) / (2*a)
        x2= (-b-math.sqrt(d)) / (2*a)
        print(f" two real and different roots are : x1:{x1} and x2: {x2}")
    else:
        print("d is not equal to zero which should be zero")
```
